## Outliers Treatment : 

```python
def outliers_detection(data):
    Q1 = np.percentile(data,25)
    Q3 = np.percentile(data,75)
    IQR = Q3 - Q1
    
    length_of_outlier =  len(data[(data < (Q1 - (1.5*IQR))) | (data > (Q3+ (1.5*IQR)))])
    return (length_of_outlier/len(data))*100
    
```

```python
def outliers_remover(data):
    for col in data.columns:
        feature = data[col]
        Q1 = np.percentile(feature,25)
        Q3 = np.percentile(feature,75)
        IQR = Q3 - Q1
        
        
        feature  = feature.loc[(feature > Q1- 1.5*IQR) & (feature < Q3 + 1.5*IQR) ]
        data = data.loc[(feature > Q1- 1.5*IQR) & (feature < Q3 + 1.5*IQR) ]
    return data
        
    

    
```

```python
def outlier_remover(a,df):

    q1 = a.quantile(.25)
    q3 = a.quantile(.75)
    iqr = q3 - q1

    maxx = q3 + 1.5 * iqr
    minn = q1 - 1.5 * iqr

    return df.loc[(a>=minn) & (a<=maxx)]
```

```python
recall = []
precision = []
fpr = []
for t in np.arange(0,1.001,0.001):
    y_prob = DTc.predict_proba(X_test)[:,1]

    y_pred_test = pd.Series(y_prob).apply(lambda i:1 if i >=t else 0 ).values
    from sklearn.metrics import confusion_matrix
    confusion_matrix = confusion_matrix(y_test,y_pred_test)
    recall.append(recall_score(y_test,y_pred_test))
    precision.append(precision_score(y_test,y_pred_test))
    TN = confusion_matrix[0][0]  # correctly predicted non-churn
    TP = confusion_matrix[-1][-1] # correctly predicted churn
    FP = confusion_matrix[0][-1] # incorrectly predicted churn
    FN = confusion_matrix[-1][0] # incorrectly predicted not-churn
    fpr_ = FP/(TN+FP)
    fpr.append(fpr_)
```
