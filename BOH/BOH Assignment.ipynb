# Applying K-Means Clustering : 

## Using cosine distance, as metric since its a very high dimentional data.

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

class KMeansCosine:
    def __init__(self, n_clusters=4, max_iter=300, random_state=None):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.random_state = random_state

    def fit(self, X):
        n_samples, n_features = X.shape

        # Initialize centroids randomly
        rng = np.random.RandomState(self.random_state)
        centroids_indices = rng.permutation(n_samples)[:self.n_clusters]
        self.centroids = X[centroids_indices]
        
        for i in range(self.max_iter):
            # Compute cosine similarities between each sample and each centroid
            cosine_similarities = cosine_similarity(X, self.centroids)

            # Assign each sample to the nearest centroid
            cluster_labels = np.argmax(cosine_similarities, axis=1)

            # Update the centroids to be the mean of the samples assigned to each cluster
            for j in range(self.n_clusters):
                mask = (cluster_labels == j)
                if np.sum(mask) > 0:
                    self.centroids[j] = np.mean(X[mask], axis=0)

    def predict(self, X):
        # Compute cosine similarities between each sample and each centroid
        cosine_similarities = cosine_similarity(X, self.centroids)

        # Assign each sample to the nearest centroid
        cluster_labels = np.argmax(cosine_similarities, axis=1)

        return cluster_labels
```

## Applying Agglomorative Clustering :  (Hierarchical Clustering)

```python
from sklearn.cluster import AgglomerativeClustering

from scipy.cluster.hierarchy import dendrogram

def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)

```

```python
Agglomerative = AgglomerativeClustering(distance_threshold=0, n_clusters=None, metric="euclidean")

Agglomerative = Agglomerative.fit(X_pca14)
plt.title("Hierarchical Clustering Dendrogram")
# plot the top three levels of the dendrogram
plot_dendrogram(Agglomerative, truncate_mode="level", p=5)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.show()
```
